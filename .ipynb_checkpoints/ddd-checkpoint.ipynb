```python
picked = int(rand(1)*100)
guess = 0
while guess != picked:
    guess = int(input("Please guess a number(1~100)>> "))
    if guess == picked:
        print("Correct!")    
    elif guess > picked:
        print("Too large, guess again>> ")
    elif guess < picked:
        print("Too small, guess again>> ")
```
