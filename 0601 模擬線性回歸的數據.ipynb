```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

準備模擬資料 f(x) = 1.2x +0.8

```python
x =np.linspace(0,5,50)
```

```python
y = 1.2*x + 0.8
```

```python
plt.scatter(x,y)
plt.plot(x,y,'r')
```

### 加入noise項 f(x) + e(x)

```python
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x,1.2*x + 0.8, 'r')
```

## 做線性迴歸

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
X = x.reshape(50,1)
```

```python
regr.fit(X, y)
```

```python
Y = regr.predict(X)
```

```python
plt.scatter(x,y)
plt.plot(x, Y, 'r')
plt.plot(x, 1.2*x + 0.8, 'g')
```
