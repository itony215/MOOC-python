```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### scikit-learn 套件

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### 讀入boston房價資訊

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

### 資料裡到底有什麼 features

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target
```

```python
len(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=87)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict)
plt.plot([0,50], [0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

### [小技巧] 善用 enumerate

```python
L = ['a', 'b', 'c']
```

```python
for i in L:
    print(i)
    
```

```python
for i in range(3):
    print(i+1, L[i])
```

```python
list(enumerate(L))
```

```python
for i in enumerate(L):
    print(i)
```

```python
for i, s in enumerate(L):
    print(i+1, s)
```

### [小技巧] 畫多個圖

```python
x = np.linspace(-10,10,200)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, x)

plt.subplot(2,2,4)
plt.plot(x, x**2)
```

### 炫炫的畫出個別參數和target關係

```python
plt.figure(figsize=(8,10))
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5,3, i+1)
    plt.scatter(X[:,i], Y, s=1)
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout()
```
